# Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('default')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x, y)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, s=75, c='green', marker='X')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y,edgecolor='black', linewidth=2, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]


plt.scatter(x, y, s=75, c=colors, cmap='Greens')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=75, c=colors, cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['ratio'].min()
```

```python
 view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidth=1,
            alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
plt.scatter(view_count, likes, edgecolors='black', linewidth=1,
            alpha=0.75)
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```

```python
plt.scatter(view_count, likes,c=ratio, edgecolors='black', linewidth=1,
            cmap='summer', alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/ Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.grid(True)
plt.show()
```
