# Histograms

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
groups = np.arange(10, 70, 10)
```

```python
plt.hist(ages, bins=groups, edgecolor='black')
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.show()
```

```python
groups = np.arange(10, 101, 10)

plt.hist(ages, bins=groups, edgecolor='black', log=True)
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.show()
```

```python
median_age = 29
color = '#fc4f30'

plt.axvline(median_age, color=color, label='Age Median', linewidth=2)
plt.hist(ages, bins=groups, edgecolor='black', log=True)
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.show()
```
